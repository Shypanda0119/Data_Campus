```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier
```

```python
X = np.load("titanic_X_train.npy")
y = np.load("titanic_y_train.npy")
```

```python
clf1 = LogisticRegression(random_state=1)
clf2 = DecisionTreeClassifier(random_state=1, max_depth=4)
eclf = VotingClassifier(estimators=[('Ir', clf1), ('rf', clf2)], voting = 'hard')
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(eclf, X, y, cv=5).mean()
```

```python
cross_val_score(clf1, X, y, cv=5).mean()
```

```python
cross_val_score(clf2, X, y, cv=5).mean()
```

```python
clf1 = LogisticRegression(random_state=1)
clf2 = DecisionTreeClassifier(random_state=1)
eclf = VotingClassifier(estimators=[('lr', clf1), ('dt', clf2)], voting='hard')
```

```python
c_params = [0.1, 5.0, 7.0, 10.0, 15.0, 20.0, 100.0]
params ={
  "lr__solver" : ['liblinear'],
  "lr__penalty" : ["l2"],
  "lr__C" : c_params,
  "dt__criterion" : ["gini", "entropy"],
  "dt__max_depth" : [10,8,7,6,5,4,3,2],
  "dt__min_samples_leaf": [1,2,3,4,5,6,7,8,9]
  #단말 노드가 되기 위한 최소한의 샘플 데이터 수
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=eclf, param_grid=params, cv=5)
grid = grid.fit(X, y)
grid.best_score_
```

```python
grid.best_params_
```

```python
from sklearn.ensemble import RandomForestClassifier
eclf =  RandomForestClassifier(n_estimators=100, max_features=2, n_jobs=-1, oob_score=True)
from sklearn.model_selection import cross_val_score
```

```python
params ={
  "n_estimators" : [10,20,30,40,50,55],
  "max_samples" : [0.5,0.6,0.7,0.8,0.9,1]
}
```

```python
grid = GridSearchCV(estimator=eclf, param_grid=params, cv=5)
grid = grid.fit(X, y)
grid.best_score_
```
