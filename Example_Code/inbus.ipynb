```python
import pandas as pd
bus_data = pd.read_excel('bus.xlsx')
bus_data.info()
```

```python
bus_data['승차정류장명']
```

```python
bus_station = bus_data['승차정류장명'].str.split(';')
bus_station
```

```python
bus_station.dropna(inplace=True)
bus_station
```

```python
exploded_bus_station = bus_station.explode()
exploded_bus_station
```

```python
size_bus_station = exploded_bus_station.groupby(exploded_bus_station).size()
size_bus_station
```

```python
size_bus_station.sort_values(ascending = False, inplace = True)
size_bus_station
```

```python
frame = {'승차정류장' : size_bus_station.index, 'count' : size_bus_station.values}
bus_station_df = pd.DataFrame(frame)
bus_station_df.head()
```

```python
bus_station_df.plot.bar(x='승차정류장', y='count', rot=90)
```

```python
top_3 = bus_station_df.nlargest(3, 'count')
top_3
```

```python
top_3.plot.bar(x='승차정류장', y='count', rot=0)
```

```python
bus_intime = bus_data[["승차정류장명","시간"]]
bus_intime
```

```python
bus_intime = bus_intime.dropna()
bus_intime
```

```python
in_bus = bus_intime[(bus_intime["시간"] <= 9)]
out_bus = bus_intime[(bus_intime["시간"] >= 17)]
inout_bus = pd.concat([in_bus, out_bus])
inout_station_counts = inout_bus['승차정류장명'].value_counts().sort_values(ascending=False)
inout_station_counts
```

```python
frame_2 = {'승차정류장' : inout_station_counts.index, 'count' : inout_station_counts.values}
inout_bus_station_df = pd.DataFrame(frame_2)
inout_bus_station_df.head()
```

```python
inout_bus_station_df.plot.bar(x='승차정류장', y='count', rot=90)
```

```python
inout_top_3 = inout_bus_station_df.nlargest(3, 'count')
inout_top_3
```

```python
inout_top_3.plot.bar(x='승차정류장', y='count', rot=0)
```
