```python
import csv

f = open('weather.csv', encoding='CP949')
data = csv.reader(f)
header = next(data)
for row in data:
  print(row[3], end=',')
f.close()
```

```python
max_wind = 0.0
for row in data:
  if row[3] == '':
    wind = 0
  else :
    wind = float(row[3])
  if max_wind < wind:
    max_wind = wind

print("지난 10년간 울릉도의 최대 풍속은 ", max_wind, 'm/s')
```

```python
import datetime as dt

weather = pd.read_csv('weather.csv', encoding='CP949')
monthly = [ None for x in range(12)]
monthly_wind = [0 for x in range(12)]
weather['month'] = pd.DatetimeIndex(weather['일시']).month

for i in range(12):
  monthly[i] = weather[weather['month'] == i + 1]
  monthly_wind[i] = monthly[i].mean()['평균 풍속(m/s)']

plt.plot(monthly_wind, 'red')
plt.show()
```

```python
weather['month'] = pd.DatetimeIndex(weather['일시']).month
means = weather.groupby('month').mean()
print(means)
```

```python
df_1 = pd.DataFrame( {'A' : ['a10', 'a11', 'a12'],
                      'B' : ['b10', 'b11', 'b12'],
                      'C' : ['c10', 'c11', 'c12']}, index = ['가', '나', '다'])

df_2 = pd.DataFrame( {'B' : ['b23', 'b24', 'b25'],
                      'C' : ['c23', 'c24', 'c25'],
                      'D' : ['d23', 'd24', 'd25']}, index = ['다', '라', '마'])
print(df_1, df_2)
```

```python
df_3 = pd.concat( [df_1, df_2])
print(df_3)
```

```python
pd.concat([df_1, df_2], axis = 0, join = 'outer')
```

```python
pd.concat([df_1, df_2], axis = 0, join = 'inner')
```

```python
pd.concat([df_1, df_2], axis = 1, join = 'outer')
```

```python
pd.concat([df_1, df_2], axis = 1, join = 'inner')
```

```python
df_3 = df_1.merge(df_2, how='outer', on='B')
print(df_3)
```

```python
df_3 = df_1.merge(df_2, how='outer', left_index = True, right_index = True)
print(df_3)
```

```python
print('left outer \n', df_1.merge(df_2, how='left', on='B'))
```

```python
print('right outer \n', df_1.merge(df_2, how='right', on='B'))
```

```python
print('full outer \n', df_1.merge(df_2, how='outer', on="B"))
```

```python
print('inner \n', df_1.merge(df_2, how='inner', on='B'))
```
