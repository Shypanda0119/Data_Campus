```python
def para_func(*para): # 매개변수(튜플형태) 여러개를 읽기위해 *para 사용,
  result = 0
  for num in para:
    result += num

  return result

hap = 0

hap = para_func(10, 20) # 호출한 매개변수는 튜플 형태로 전달
print("매개변수가 2개인 함수를 호출한 결과:%d" % hap)
hap = para_func(10, 20, 30) # 호출한 매개변수는 튜플 형태로 전달
print("매개변수가 3개인 함수를 호출한 결과:%d" % hap)
```

```python
def dic_func(**para): # 매개변수(딕셔너리 형태) 여러개를 읽기위해 **para 사용 (키 = 값 형식)
  for k in para.keys():
    print("%s, --> %d명입니다." % (k,para[k]))

dic_func(트와이스 = 9, 소녀시대 = 7, 걸스데이 = 4, 에이핑크 = 6)
```

```python
def print_baggage_items(*items, **item_with_count):
  print("# 단일품목")
  for item in items:
      print(item[0])
      for subitem in item[1:]:
          print(subitem)
  print()
  print("# 다중품목")
  for num in item_with_count.keys():
    print("%s, --> %d 개" % (num,item_with_count[num]))
print_baggage_items(("laptop", "camera", "charger"),socks = 8,pants = 2,shirts = 4)
```

```python
def print_country(country, code, alpha3, alpha2):
    print(country, code, alpha3, alpha2)

def print_all_countries(countries):
    for country_info in countries:
        print_country(**country_info)

countries = (
    {'country': '뉴질랜드', 'code': 554, 'alpha3': 'NZL', 'alpha2': 'NZ'},
    {'country': '대한민국', 'code': 410, 'alpha3': 'KOR', 'alpha2': 'KR'},
    {'country': '미국', 'code': 840, 'alpha3': 'USA', 'alpha2': 'US'},
)

print_all_countries(countries)
```

```python
num_students = int(input("입력하는 학생수가 총 몇명인가요? : "))
print("학생의 이름과 시험 점수를 차례대로 입력하세요!")
data = []
for x in range(1, num_students + 1):
  print("%d 번째 학생 ==" % x)
  students_name = input("* 이름: ")
  students_score = int(input("* 점수: "))
  students_data = (students_name, students_score)
  data.append(students_data)

while True:
  who_score = input("어떤 학생의 점수가 궁금하신가요? 이름을 입력하세요. : ")
  found = False
  for students_data in data:
    if students_data[0] == who_score:
      print(who_score,"학생의 점수 : %d" % students_data[1])
      found = True
      break
  if not found:
    print("찾는 학생(", who_score, ")의 점수가 존재하지 않습니다.")
  else:
    print('프로그램을 종료합니다.')
    break
```

```python
children = ["안나", "유나", "신후", "원준"]
children_fight = []
for a in range(len(children)):
  for b in range(len(children)):
    if a != b:
      children_fight.append((children[a], children[b]))
print(children_fight)
```
