```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

seoul = pd.read_excel('weather.xlsx')
seoul.head()
```

```python
seoul.info()
```

```python
seoul_new = seoul[['날짜', '평균기온(℃)']]
```

```python
seoul_new = seoul.rename(columns = {'날짜':'date', '평균기온(℃)':'temperature'})
seoul_new['year'] = seoul_new['date'].dt.year
seoul_new = seoul_new[['year','temperature']]
```

```python
seoul_new
```

```python
year_temperature = seoul_new.groupby("year")['temperature'].agg([('평균기온', 'mean')]).reset_index()
year_temperature
```

```python
plt.plot(year_temperature['year'], year_temperature["평균기온"], marker='o')
plt.xlabel('년도')
plt.ylabel('평균기온(℃)')
plt.title('년도별 평균 기온')
plt.grid(True)
plt.show()
```

```python
X = seoul_new[['year']]
y = seoul_new["temperature"]
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
prediction = model.predict([[2022]])
print(f"2022년 평균 기온 예측: {prediction[0]}")
```

```python
X = seoul[["날짜", '최저기온(℃)', '최고기온(℃)']]
y = seoul['평균기온(℃)']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, test_size = 0.3)
model.fit(X_train, y_train)
plt.figure(figsize=(15,7))
plt.plot(model.predict(X_test[:50]), label='predict')
plt.plot(y_test[:50].values.reshpae(-1, 1), label = 'real temp')
plt.legend()
```
