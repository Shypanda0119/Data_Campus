```python
#matmatplotlib의 폰트를 Nanum 폰트로 지정
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')
```

```python
import pandas as pd
data = pd.read_excel('wild boar.xlsx')
```

```python
grouped_data = data.groupby('시군')['개체수'].sum()
grouped_data
```

```python
area_data=data.groupby('시군')[['경도','위도']].mean()
area_data
```

```python
merge_data = pd.merge(grouped_data, area_data, on='시군')
merge_data
```

```python
top_7 = merge_data.nlargest(7, '개체수')
top_7
```

```python
import folium
map_data =  folium.Map(location=[36.30, 126.98], zoom_start = 12)
```

```python
from folium.plugins import HeatMap
HeatMap(data = top_7[['경도','위도','개체수']]).add_to(map_data)
map_data

```

#2-1 연도별 포획 실적

```python
grouped_year_data = data.groupby([data['포획일'].dt.year, '시군'])['개체수'].sum()
grouped_year_data
```

```python
import matplotlib.pyplot as plt

# 그래프 크기 설정
#plt.figure(figsize=(12, 8)) 값을 바꿔도 그래프가 커지 않아 다른코드 사용
plt.rcParams["figure.figsize"] = (15, 10)

# 막대 그래프 그리기
ax = grouped_year_data.unstack().plot(kind='bar')

# 그래프 제목 및 축 이름 설정
plt.title('연도별 개체수')
plt.xlabel('연도')
plt.ylabel('개체수')

# 범례 표시
plt.legend(fontsize='small')

# x축 눈금 레이블 회전(90도: rotation=90)
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)

# 그래프 출력
plt.show()
```

#2-2 월별 포획 실적

```python
grouped_month_data = data.groupby([data['포획일'].dt.month, '시군'])['개체수'].sum()
grouped_month_data
```

```python
# 그래프 크기 설정
plt.rcParams["figure.figsize"] = (17, 10)

# 막대 그래프 그리기
ax = grouped_month_data.unstack().plot(kind='bar')

# 그래프 제목 및 축 이름 설정
plt.title('월별 개체수')
plt.xlabel('월')
plt.ylabel('개체수')

# 범례 표시
plt.legend(fontsize='small')

# x축 눈금 레이블 회전(90도: rotation=90)
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)

# 그래프 출력
plt.show()
```

#2-3 계절별 포획 실적

```python
# 데이터셋에서 필요한 열을 선택하여 데이터프레임으로 생성
selected_data = data[['포획일', '시군', '개체수']]

# 계절을 나타내는 열 생성
selected_data['계절'] = selected_data['포획일'].dt.month.map({
    12: '겨울', 1: '겨울', 2: '겨울',
    3: '봄', 4: '봄', 5: '봄',
    6: '여름', 7: '여름', 8: '여름',
    9: '가을', 10: '가을', 11: '가을'
})

# 계절별 포획실적 그룹화
grouped_season_data = selected_data.groupby(['계절', '시군'])['개체수'].sum()

# 그래프 그리기
plt.figure(figsize=(10, 6))
ax = grouped_season_data.unstack().plot(kind='bar', width=0.6)
plt.title('계절별 포획실적')
plt.xlabel('계절')
plt.ylabel('개체수')
plt.legend(fontsize='large')

# x축 눈금 레이블 회전(90도: rotation=90)
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)

plt.show()
```
